# Import libraries

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

# Read csv file

```python
data = pd.read_csv("HDFCBANK_01-01-2021_27-09-2021.csv")
```

```python
#head returns 5 rows by default if we want more columns use data.head(number_of_rows_to_display) example: data.head(10)
#which display 10 rows.
data.head()
```

# Rename the columns as
    Date -> date
    High Price -> high
    Low Price -> low
    Total Trade -> trade
    Turnover (in lacs) -> turnover_in_lakhs
    No. of Contracts -> contracts


```python
#create a dictionary to rename the columns

dict1 = {'Date':'date',
        'High Price':'high',
        'Low Price':'low',
        'Total Trade': 'trade',
        'Turnover (in lacs)':'turnover_in_lakhs',
        'No. of Contracts':'contracts'}

#rename the columns using rename function 
data.rename(columns = dict1, inplace = True)

#Display data
#head returns 5 rows by default if we want more columns use data.head(number_of_rows_to_display) example: data.head(10)
#which display 10 rows
data.head()
```

# Add a new column name “difference”
    
   	Difference for each row will be [high - low]


```python
#data['column_name'] -> selecting column from the dataset

difference = data['high']-data['low']

# add difference column into dataframe
data['difference'] = difference

#display data
#head returns 5 rows by default if we want more columns use data.head(number_of_rows_to_display) example: data.head(10)
#which display 10 rows
data.head()
```

# Save the data into file named “processed_data.csv”

```python
# save the data by using to_csv function
data.to_csv('processed_data.csv')
```
